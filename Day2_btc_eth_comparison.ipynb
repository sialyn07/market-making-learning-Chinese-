```python
import ccxt

coinbase = ccxt.coinbase()
pairs = ['BTC/USDT', 'ETH/USDT']

for pair in pairs:
    print(f"\n{'='*40}")
    print(f"  {pair}")
    print(f"{'='*40}")
    ob = coinbase.fetch_order_book(pair, limit=5)
    bid = ob['bids'][0][0]
    ask = ob['asks'][0][0]
    spread = ask - bid
    spread_bp = (spread / bid) * 10000
    bid_depth = sum(b[1] for b in ob['bids'])
    ask_depth = sum(a[1] for a in ob['asks'])
    print(f"  最优买价: ${bid:.2f} | 最优卖价: ${ask:.2f}")
    print(f"  价差: ${spread:.2f} ({spread_bp:.2f} bp)")
    print(f"  前5档买单深度: {bid_depth:.4f} | 卖单深度: {ask_depth:.4f}")
```
