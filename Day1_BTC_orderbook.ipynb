```python
import ccxt

coinbase = ccxt.coinbase()
ob = coinbase.fetch_order_book('BTC/USDT', limit=5)

bid = ob['bids'][0][0]
ask = ob['asks'][0][0]
spread = ask - bid
spread_bp = (spread / bid) * 10000
bid_depth = sum(b[1] for b in ob['bids'])
ask_depth = sum(a[1] for a in ob['asks'])

print(f'жңҖдјҳд№°д»·: ${bid:.2f}')
print(f'жңҖдјҳеҚ–д»·: ${ask:.2f}')
print(f'д»·е·®: ${spread:.2f} ({spread_bp:.2f} bp)')
print(f'еүҚ5жЎЈд№°еҚ•ж·ұеәҰ: {bid_depth:.4f} BTC')
print(f'еүҚ5жЎЈеҚ–еҚ•ж·ұеәҰ: {ask_depth:.4f} BTC')
```
