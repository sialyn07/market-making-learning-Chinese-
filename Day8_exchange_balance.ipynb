```python
# Day 8：交易所BTC储备查询（CoinGecko API）
# 不用登录，不用API key，直接跑

import requests

# 1. 十大交易所BTC交易量排名
url = "https://api.coingecko.com/api/v3/exchanges?per_page=10&page=1"
resp = requests.get(url)
exchanges = resp.json()

print("=== 十大交易所BTC交易量排名 ===")
for i, ex in enumerate(exchanges, 1):
    name = ex['name']
    volume = ex['trade_volume_24h_btc']
    print(f"{i}. {name:15s} 24h交易量: {volume:>12.2f} BTC")

# 2. BTC当前价格
print("\n=== BTC当前行情 ===")
ticker_url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd&include_24hr_change=true"
ticker = requests.get(ticker_url).json()

btc_price = ticker['bitcoin']['usd']
btc_change = ticker['bitcoin']['usd_24h_change']
print(f"BTC价格: ${btc_price:,.2f}")
print(f"24h变化: {btc_change:+.2f}%")

# 3. CoinGlass数据交叉验证
# Coinbase 7天流入 2,911 BTC（从CoinGlass查到的）
inflow_7d = 2911
inflow_value = inflow_7d * btc_price
print(f"\n=== CoinGlass交叉验证 ===")
print(f"Coinbase 7天净流入: {inflow_7d:,} BTC")
print(f"约合价值: ${inflow_value:,.0f}")
```
