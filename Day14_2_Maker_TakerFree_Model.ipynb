```python
import csv
from datetime import datetime

def calc_maker_taker_pnl(bid_price, ask_price, size, maker_fee=-0.0001, taker_fee=0.0004):
    buy_cost = bid_price * size
    buy_net = buy_cost * (1 + maker_fee)
    sell_proceed = ask_price * size
    sell_net = sell_proceed * (1 - maker_fee)
    spread_pnl = (ask_price - bid_price) * size
    maker_rebate_buy = -buy_cost * maker_fee
    maker_rebate_sell = -sell_proceed * maker_fee
    net_pnl = spread_pnl + maker_rebate_buy + maker_rebate_sell
    return {
        'spread_pnl': round(spread_pnl, 2),
        'net_pnl': round(net_pnl, 2),
        'rebate_total': round(maker_rebate_buy + maker_rebate_sell, 2),
    }

def analyze_spoofing_risk():
    print("== 挂假单成本分析 ==")
    print("假设挂假单200次, 每次10 BTC, 停留3秒")
    print()
    eaten = 200 * 0.01
    cost_per = 63000 * 10 * 0.0004
    print(f"[风险1] 被吃概率1% -> 预期被吃 {eaten:.0f} 次")
    print(f"        单次被吃损失 ${cost_per:.2f}")
    print(f"        总损失 ${eaten * cost_per:.2f}")
    print()
    print(f"[风险2] 挂撤比 200:1 (阈值100:1)")
    print(f"        超阈值! 可能被API限速/涨费率/封号")
    print()
    print("[风险3] 假单挂在卖一, 来利好消息拉升,")
    print("        假单没撤掉被低价吃了 -> 亏")
    print()
    print("[风险4] Spoofing在美国是联邦罪行")
    print("        摩根大通交易员被罚$920M")

result = calc_maker_taker_pnl(63000, 63050, 1.0)
print(f"价差收益: ${result['spread_pnl']}")
print(f"返佣总额: ${result['rebate_total']}")
print(f"净收益:   ${result['net_pnl']}")
print()
print(f"修正前: $37.39  修正后: ${result['net_pnl']}")
print()

analyze_spoofing_risk()
```
