```python
import ccxt

coinbase = ccxt.coinbase()

def simulate_slippage(pair, qty):
    ob = coinbase.fetch_order_book(pair, limit=10)
    asks = ob['asks']
    filled = 0
    total_cost = 0
    line = "-" * 40
    print(f"\n  买入 {qty} {pair.split('/')[0]} 的滑点模拟：")
    print(f"  {'档位':>4} {'价格':>12} {'数量':>12} {'累计':>12} {'成本':>12}")
    print(f"  {line}")
    for i, (price, size) in enumerate(asks):
        take = min(size, qty - filled)
        filled += take
        total_cost += take * price
        print(f"  {i+1:>4} ${price:>8.2f} {take:>10.4f} {filled:>10.4f} ${total_cost:>8.2f}")
        if filled >= qty:
            break
    avg_price = total_cost / filled
    best_ask = asks[0][0]
    slip = avg_price - best_ask
    slip_pct = (slip / best_ask) * 100
    print(f"\n  最优卖价: ${best_ask:.2f}")
    print(f"  实际均价: ${avg_price:.2f}")
    print(f"  滑点: ${slip:.2f} ({slip_pct:.4f}%)")

print("=" * 40)
print("  BTC 滑点模拟")
print("=" * 40)
simulate_slippage('BTC/USDT', 0.05)

print("\n" + "=" * 40)
print("  ETH 滑点模拟")
print("=" * 40)
simulate_slippage('ETH/USDT', 0.5)
```
